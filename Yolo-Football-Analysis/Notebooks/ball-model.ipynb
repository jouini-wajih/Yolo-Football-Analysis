```python
import cv2
import random

def extract_random_frame(video_path):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Get the total number of frames in the video
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if total_frames == 0:
        print("Error: Video contains no frames.")
        return None

    # Select a random frame index
    random_frame_index = random.randint(0, total_frames - 1)

    # Set the video position to the random frame
    cap.set(cv2.CAP_PROP_POS_FRAMES, random_frame_index)

    # Read the random frame
    ret, frame = cap.read()

    # Check if frame reading was successful
    if ret:
        print(f"Random frame at index {random_frame_index} extracted.")
        cap.release()  # Release the video capture
        return frame  # Return the frame
    else:
        print(f"Error: Could not read frame at index {random_frame_index}.")
        cap.release()
        return None

# Example usage
random_frame = extract_random_frame(video_path)

# If you want to save the frame
if random_frame is not None:
    cv2.imwrite('/kaggle/working/random_frame.jpg', random_frame)
    print("Random frame saved as 'random_frame.jpg'.")
```

```python
import cv2
import random

def extract_random_frame(video_path, output_image_path):
    # Open the video file
    video_capture = cv2.VideoCapture(video_path)

    # Check if the video was opened successfully
    if not video_capture.isOpened():
        print("Error: Could not open video.")
        return

    # Get the total number of frames in the video
    total_frames = int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # Select a random frame number
    random_frame_number = random.randint(0, total_frames - 1)

    # Set the video capture to the random frame
    video_capture.set(cv2.CAP_PROP_POS_FRAMES, random_frame_number)

    # Read the frame
    success, frame = video_capture.read()
    
    # Check if the frame was read successfully
    if not success:
        print("Error: Could not read frame.")
        return

    # Save the extracted frame as an image
    cv2.imwrite(output_image_path, frame)

    # Release the video capture object
    video_capture.release()

    print(f"Extracted frame {random_frame_number} and saved as {output_image_path}")

# Example usage
video_path = "/kaggle/input/-dfl-bundesliga-460-mp4-videos-in-30sec-csv/DFL Bundesliga Data Shootout/test/test (1).mp4"
output_image_path = "random_frame.jpg"  # Specify the output image file name

extract_random_frame(video_path, output_image_path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt  # Import for displaying images
from ultralytics import YOLO  # For YOLOv8

# Load the YOLO model
model = YOLO('/kaggle/working/ball_models/ball_model_v06/weights/best.pt')

def process_image_grid(image_path, grid_size=(6, 6)):
    # Load the image
    img = cv2.imread(image_path)
    img_height, img_width, _ = img.shape
    
    # Determine the size of each grid cell
    cell_width = img_width // grid_size[1]
    cell_height = img_height // grid_size[0]
    
    # List to hold annotations for detected objects
    annotations = []
    
    # Iterate through the grid
    for i in range(grid_size[0]):  # Row index
        for j in range(grid_size[1]):  # Column index
            # Calculate grid cell boundaries
            start_x = j * cell_width
            start_y = i * cell_height
            end_x = start_x + cell_width
            end_y = start_y + cell_height
            
            # Extract the grid cell from the image
            grid_cell = img[start_y:end_y, start_x:end_x]
            
            # Pass the grid cell to YOLO for object detection
            results = model(grid_cell)
            
            # Annotate detected objects within the grid cell
            for result in results:
                for box in result.boxes:
                    # Bounding box coordinates relative to the grid cell
                    x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)
                    label = result.names[int(box.cls)]
                    confidence = box.conf.item()  # Convert tensor to Python float
                    
                    # Adjust bounding box coordinates relative to the original image
                    x1_full = start_x + x1
                    y1_full = start_y + y1
                    x2_full = start_x + x2
                    y2_full = start_y + y2
                    
                    # Append the bounding box and label to annotations
                    annotations.append((x1_full, y1_full, x2_full, y2_full, label, confidence))
                    
                    # Draw the bounding box on the original image
                    cv2.rectangle(img, (x1_full, y1_full), (x2_full, y2_full), (0, 255, 0), 2)
                    cv2.putText(img, f"{label} {confidence:.2f}", (x1_full, y1_full - 10),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    
    # Save or return the final annotated image
    output_image_path = "annotated_image.jpg"
    cv2.imwrite(output_image_path, img)
    
    # Display the annotated image
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Hide axes
    plt.show()
    
    return output_image_path, annotations

# Example usage
image_path = "/kaggle/working/random_frame.jpg"  # Replace with your image path
output_img, detected_annotations = process_image_grid(image_path)

print(f"Output image saved at: {output_img}")
print(f"Detected annotations: {detected_annotations}")
```

```python
import cv2
import numpy as np
from ultralytics import YOLO  # For YOLOv8
import os

# Load the YOLO model
model = YOLO('/kaggle/working/ball_models/ball_model_v06/weights/best.pt')

def process_image_grid(img, grid_size=(6, 6)):
    img_height, img_width, _ = img.shape
    
    # Determine the size of each grid cell
    cell_width = img_width // grid_size[1]
    cell_height = img_height // grid_size[0]
    
    # List to hold annotations for detected objects
    annotations = []
    
    # Iterate through the grid
    for i in range(grid_size[0]):  # Row index
        for j in range(grid_size[1]):  # Column index
            # Calculate grid cell boundaries
            start_x = j * cell_width
            start_y = i * cell_height
            end_x = start_x + cell_width
            end_y = start_y + cell_height
            
            # Extract the grid cell from the image
            grid_cell = img[start_y:end_y, start_x:end_x]
            
            # Pass the grid cell to YOLO for object detection
            results = model(grid_cell)
            
            # Annotate detected objects within the grid cell
            for result in results:
                for box in result.boxes:
                    # Bounding box coordinates relative to the grid cell
                    x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)
                    label = result.names[int(box.cls)]
                    confidence = box.conf.item()  # Convert tensor to Python float
                    
                    # Adjust bounding box coordinates relative to the original image
                    x1_full = start_x + x1
                    y1_full = start_y + y1
                    x2_full = start_x + x2
                    y2_full = start_y + y2
                    
                    # Append the bounding box and label to annotations
                    annotations.append((x1_full, y1_full, x2_full, y2_full, label, confidence))
                    
                    # Draw the bounding box on the original image
                    cv2.rectangle(img, (x1_full, y1_full), (x2_full, y2_full), (0, 255, 0), 2)
                    cv2.putText(img, f"{label} {confidence:.2f}", (x1_full, y1_full - 10),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    return img, annotations

def process_video(video_path, output_path):
    # Open the video file
    video_capture = cv2.VideoCapture(video_path)
    
    # Get video properties
    fps = int(video_capture.get(cv2.CAP_PROP_FPS))
    width = int(video_capture.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(video_capture.get(cv2.CAP_PROP_FRAME_HEIGHT))
    total_frames = int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT))

    # Create a VideoWriter to save the processed video
    fourcc = cv2.VideoWriter_fourcc(*'XVID')  # Specify the codec
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    # Process each frame
    for frame_num in range(total_frames):
        ret, frame = video_capture.read()
        if not ret:
            print(f"Error: Could not read frame {frame_num}.")
            break

        # Process the frame through the YOLO model
        processed_frame, annotations = process_image_grid(frame)
        
        # Write the processed frame to the output video
        out.write(processed_frame)

        # Print progress
        print(f"Processed frame {frame_num + 1}/{total_frames}")

    # Release resources
    video_capture.release()
    out.release()
    print(f"Output video saved at: {output_path}")

# Example usage
video_path = "/kaggle/input/-dfl-bundesliga-460-mp4-videos-in-30sec-csv/DFL Bundesliga Data Shootout/test/test (1).mp4"
output_path = "output_video.avi"  # Specify the output video file name

process_video(video_path, output_path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random

def plot_random_frames(video_path, num_frames=5):
    # Open the video file
    video_capture = cv2.VideoCapture(video_path)

    # Get total number of frames in the video
    total_frames = int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # Generate random frame indices
    random_indices = random.sample(range(total_frames), num_frames)
    random_indices.sort()  # Sort for sequential reading

    # List to hold the frames
    frames = []

    # Read and store the random frames
    for index in random_indices:
        video_capture.set(cv2.CAP_PROP_POS_FRAMES, index)  # Set the frame position
        ret, frame = video_capture.read()
        if ret:
            # Convert BGR to RGB format for displaying
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frames.append(frame_rgb)
        else:
            print(f"Error: Could not read frame {index}.")
            break

    # Release video capture object
    video_capture.release()

    # Plot the frames
    plt.figure(figsize=(15, 10))
    for i, frame in enumerate(frames):
        plt.imshow(frame)
        plt.axis('off')  # Hide axes
    plt.tight_layout()
    plt.show()

# Example usage
video_path = "/kaggle/working/output_video.avi"
num_frames_to_plot = 5  # Specify how many random frames to plot

plot_random_frames(video_path, num_frames_to_plot)
```
